#### Challenge 1
#### 1.- Define a function that simulates rolling a dice 10 times. Save the information in a dataframe.
**Hint**: you can use the *choices* function from module *random* to help you with the simulation.

```python
# your code here

def rolling_dice_sim():
     # Simulating 10 dice rolls; a die has 6 faces (1-6)
    rolls = random.choices(range(1,7),k=10)
    # Creating a dataframe to store the results
    df = pd.DataFrame(rolls, index=range(1,11), columns=['Dice Result'])

    return df
# Running the function to see the results: 

df = rolling_dice_sim()

display(df)
```

#### 2.- Plot the results sorted by value.

```python
# your code here
 

# Sort the dataframe by 'Dice Rolls' and reset index
sorted_df = df.sort_values(by='Dice Result').reset_index(drop=True)
# change index to start at 1
sorted_df.index = sorted_df.index + 1

# Plotting the sorted results
plt.scatter(sorted_df.index, sorted_df['Dice Result'], color='red')  # Create a scatter plot, I first chose bar plot but with scatter it is easier to see the dice result for each roll
plt.xlabel('Dice Rolls')  # Label for x-axis
plt.ylabel('Dice Roll Outcome')  # Label for y-axis
plt.title('Dice Rolls Sorted by Outcome')  # Title of the plot
plt.xticks(range(1, 11))  # Set x-ticks to show original roll numbers
plt.grid(True) # add a grid for readability
plt.show()
```

#### 3.- Calculate the frequency distribution and plot it. What is the relation between this plot and the plot above? Describe it with words.

```python
# your code here
# Calculating the frequency distribution of the dice rolls
frequency = df['Dice Result'].value_counts().sort_index()
frequency
# plotting the frequency results: 

plt.bar(frequency.index, frequency,color='purple') # create a bar plot using frequency data
plt.xlabel('Dice Roll Outcome') # setting x label as the roll results (numbers in the dice)
plt.ylabel('Frequency') # setting y label as the fequency of each number
plt.title('Frequency distribution of dice rolls') # setting a title for the plot
plt.xticks(range(1, 7))  # Set x-ticks to show all possible dice roll outcomes

plt.show()
```

```python
"""
While the plot of sorted values gives a detailed view of each roll's outcome in a specific order, 
the frequency plot provides an overview of the overall outcomes, summarizing the entire set of rolls.
The sorted values plot  displays each dice roll's outcome in a visually sorted order.
Each point or bar represents a specific dice roll, showing how each outcome compares to the others when arranged sequentially.
The frequency plot aggregates the results to show how many times each possible outcome occurred across all rolls. 
It summarizes the data by counting occurrences rather than showing each roll individually, as the plot of sorted values.
Providing complementary perspectives, together both plots allow us to see a fuller picture of the data's characteristics than either could alone.
"""
```

Now, using the dice results obtained in *challenge 1*, your are going to define some functions that will help you calculate the mean of your data in two different ways, the median and the four quartiles. 

#### 1.- Define a function that computes the mean by summing all the observations and dividing by the total number of observations. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
# your code here
def mean_calculation(column):
    """
    Takes a column with numerical values that can be summed and counted, sums and counts them without using built-in functions, 
    then divides the sum by the count, and returns the mean of the data provided
    """
     #Check if the data list is not empty
    if len(column) == 0:
        return None # handling empty data
    
    total_sum = 0  # Initialize a variable to store the sum of all elements
    count = 0      # Initialize a counter to count the number of elements
    
    # Loop through each element in the data list
    for value in column:
        total_sum += value  # Add each element to total_sum
        count += 1          # Increment the count for each element
    
    # Calculate the mean by dividing the total sum by the number of elements
    mean = total_sum / count
    
    return mean

mean_calculation(df['Dice Result'])
    
```

#### 2.- First, calculate the frequency distribution. Then, calculate the mean using the values of the frequency distribution you've just computed. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
def calculate_mean_from_frequency(data):
    # Calculating the frequency distribution using another method
    frequency_distribution = {} # create an empty dictionary
    for number in data:
        if number in frequency_distribution:
            frequency_distribution[number] += 1
        else:
            frequency_distribution[number] = 1
    # Calculating the mean using the frequency distribution
    total_sum = 0 # start the sum at 0
    total_count = 0 # start the count at 0
    for number, frequency in frequency_distribution.items():
        total_sum += number * frequency # sum it dice value multiplied by it's frequency
        total_count += frequency  # sum the frequencies

    # Calculating the mean
    mean = total_sum / total_count
    return mean

calculate_mean_from_frequency(df['Dice Result'])
```

#### 3.- Define a function to calculate the median. You are not allowed to use any methods or functions that directly calculate the median value. 
**Hint**: you might need to define two computation cases depending on the number of observations used to calculate the median.

```python
# your code here
def calculate_median(data):
    # Step 1: Sort the data
    sorted_data = sorted(data)
    
    # Step 2: Determine the number of observations
    n = len(sorted_data)
    
    # Step 3: Compute the median
    if n % 2 == 1:
        # If odd, the median is the middle element
        median = sorted_data[n // 2]
    else:
        # If even, the median is the average of the two middle elements
        mid_index = n // 2
        median = (sorted_data[mid_index - 1] + sorted_data[mid_index]) / 2
    
    return median
```

#### 4.- Define a function to calculate the four quartiles. You can use the function you defined above to compute the median but you are not allowed to use any methods or functions that directly calculate the quartiles. 

```python
# your code here

def calculate_quartiles(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    # Calculate Q2 (Median)
    Q2 = calculate_median(sorted_data)

    # Find the split index for odd and even length datasets
    if n % 2 == 0:
        lower_half = sorted_data[:n // 2]
        upper_half = sorted_data[n // 2:]
    else:
        lower_half = sorted_data[:n // 2]
        upper_half = sorted_data[n // 2 + 1:]

    # Calculate Q1 and Q3 using the median function on halves
    Q1 = calculate_median(lower_half)
    Q3 = calculate_median(upper_half)
    Q4 = sorted_data[-1]  # Maximum value

    return Q1, Q2, Q3, Q4
```
